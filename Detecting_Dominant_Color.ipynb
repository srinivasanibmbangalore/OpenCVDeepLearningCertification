```
import numpy as np
import matplotlib.pyplot as plt
import cv2
import matplotlib
```

```
from google.colab import drive

drive.mount('/content/gdrive')
root_path = '/content/gdrive/My Drive/ml_data/OpenCVCertification/images/'
vegPath=root_path+"/india cricket t shirt.jpeg"
```

```
image= cv2.imread(vegPath, 1)
if ( image.any() == None):
    print("file not detected")
else:
    print("file detected")
matplotlib.rcParams['figure.figsize'] = (10.0, 10.0)
```

```
plt.imshow(image[:,:,::-1])
plt.title("Original Image")
```

```
# convert from bgr to HSV format
hsvImage = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
H,S,V=cv2.split(hsvImage)
```

```
H_array = H[S>10].flatten()  # Removes the white background
```

```
plt.figure(figsize=[20,10])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Image");plt.axis('off')
plt.subplot(122);plt.hist(H_array, bins=180, color='r');plt.title("Histogram")
```

From the histogram plot, we can see that the peaks are at 20 (i.e. hue = 46, Orange) , 122 (i.e. hue 240, blue-ish) . These correspond to Red and Blue Colors. The highest peak is for blue since that is the most dominant color in the image.
